```python
import os
import pandas as pd
```

```python
all_images = os.listdir("/Users/deveshshah/Documents/Programming/Out_Of_Distribution_Exploration/dataset/imagenet-o/images_resized")
```

```python
len(all_images)
```

```python
all_images = [x for x in all_images if x.endswith(".jpg")]
```

```python
len(all_images)
```

```python
df = pd.DataFrame(all_images, columns=["image_path"])
df["stage"] = "train"
df["label"] = "ood_wild"
df["id"] = "imagenet-o_" + df.index.astype(str)
```

```python
df = df[["id", "image_path", "label", "stage"]]
```

```python
df.head()
```

```python
# randomly set 20% of data to val
from random import random, sample
all_ids = df["id"].tolist()
val_ids = set(sample(all_ids, int(0.2 * len(all_ids))))
df["stage"] = df["id"].apply(lambda x: "val" if x in val_ids else "train")
```
